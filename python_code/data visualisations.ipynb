```python
#importing the libraries needed to create visualisations and statistical analysis

import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
import seaborn as sns
from scipy import stats
from matplotlib import rcParams
```

```python
### RATINGS RATINGS RATINGS
```

```python
# calling the data on film-revenue vs. film-ratings so that it can later be plotted
data = "./data/revenue_ratings/movies_revenue_ratings_new copy.csv"
data_plot = pd.read_csv(data, index_col=0)

# testing the data has been called
data_plot.head()
```

```python
# plotting the scattergraph with line of best fit for film weighted-rating vs. film revenue
rating1 = sns.regplot(x="weighted_rating", y="revenue", data=data_plot, line_kws={'color': 'red'})

# setting the size of the graph
sns.mpl.rc("figure", figsize=(8, 6))

# plotting graph title and axis labels
plt.title('Comparing film ratings and film revenues')
plt.xlabel('film rating')
plt.ylabel('film revenue: ($billions)')

#saving the figure
rating1.figure.savefig("rating1.png")
```

```python
# calling data on film revenue vs. film rating, where rating has been rounded to 0 decimal places 
revenue2 = "./data/revenue_ratings/movies_revenue_ratings_new.csv"
revenue_2 = pd.read_csv(revenue2, index_col=0)

# testing the data has been called
revenue_2.head()
```

```python
#setting the grid for the graph
sns.set_style('whitegrid')
 
# Violin plot of film revenue vs. film rating
rating2 = sns.violinplot(x='weighted_rating', y='revenue', data=revenue_2)

# setting the graph size
sns.mpl.rc("figure", figsize=(20, 10))

# plotting graph title and axis labels
plt.title('Comparing the distribution of film revenues within film ratings')
plt.xlabel('film rating')
plt.ylabel('film revenue: ($billions)')

# saving the figure
rating2.figure.savefig("ratingviolin.png")
```

```python
# Bar graph of film revenue vs. film rating
az = sns.barplot(x="weighted_rating", y="revenue", data=data_plot, capsize=.2)

# setting the graph size
sns.mpl.rc("figure", figsize=(8, 6))

# plotting graph title and axis labels
plt.title('Comparing the distribution of film ratings and film revenue')
plt.xlabel('film rating')
plt.ylabel('film revenue: ($billions)')
plt.xticks(rotation=-90)

# saving the figure
az.figure.savefig("ratingbarchart1.png")
```

```python
# Bar graph of film revenue vs. film rating (film ratings have been rounded to 0 decimal places)
aa = sns.barplot(x="weighted_rating", y="revenue", data=revenue_2, capsize=.2)

# setting the graph size
sns.mpl.rc("figure", figsize=(8, 6))

# plotting graph title and axis labels
plt.title('Comparing the distribution of film ratings and film revenue')
plt.xlabel('film rating')
plt.ylabel('film revenue: ($billions)')
plt.xticks(rotation=-90)

# saving the figure
aa.figure.savefig("ratingbarchart2.png")
```

```python
#BUDGET BUDGET BUDGET
```

```python
# calling the data on film-revenue vs. film-budget so that it can later be plotted
budget = "./data/budget/budget_1.csv"
budget1 = pd.read_csv(budget, index_col=0)

# testing the data has been called
budget1.head()
```

```python
#scatterplot for film budget vs. film revenue
sns.set(style="white")
ac = sns.lmplot(data=budget1, x="budget", y="revenue", fit_reg=True, line_kws={'color': 'red'})

#setting figure size
sns.mpl.rc("figure", figsize=(12, 10))

# plotting graph title and axis labels
plt.title('Comparing film ratings and film budget')
plt.xlabel('film budget: (1 = $100 million)')
plt.ylabel('film revenue: ($billions)')
plt.xticks(rotation=-90)

#saving the figure
ac.savefig("budgetreg.png")
```

```python
#rounding the budget so that similar budgets can group together
budget1.budget = budget1.budget.round(-7)

# testing the budget has been rounded
budget1.head()
```

```python
# bar graph for film budget vs. film revenue
ax = sns.barplot(x="budget", y="revenue", data=budget1)

#setting figure size
sns.mpl.rc("figure", figsize=(10, 8))

# plotting graph title and axis labels
plt.title('Comparing the distribution of film revenue and film budget')
plt.xlabel('film budget ($)')
plt.ylabel('film revenue: ($billions)')
plt.xticks(rotation=-90)

#saving the figure
ax.figure.savefig("budgetbarchart2.png")
```

```python
# violin graph for film budget vs. film revenue
hb = sns.violinplot(x='budget', y='revenue', data=budget1)

#setting figure size
sns.mpl.rc("figure", figsize=(12, 8))

# plotting graph title and axis labels
plt.title('Comparing the distribution of film revenue within film budgets')
plt.xlabel('film budget ($)')
plt.ylabel('film revenue: ($billions)')
plt.xticks(rotation=-90)

#saving the figure
hb.figure.savefig("budgetviolingraph.png")
```

```python
#calling data on film revenue and film gender-split
gender = "./data/actors_revenue/movies_gender.csv"
gender1 = pd.read_csv(gender, index_col=0)

# testing the data has been called
gender1.head()
```

```python
#scatterplot for film gender-split vs. film revenue (male actors)
fg = sns.regplot(x="percentage male", y="revenue", data=gender1, line_kws={'color': 'red'})

#setting figure size
sns.mpl.rc("figure", figsize=(8, 6))

# plotting graph title and axis labels
plt.title('Comparing film revenue and the percentage of men in films')
plt.xlabel('percentage of film actors that are male: (1.0 = 100%)')
plt.ylabel('film revenue ($billions)')
plt.xticks(rotation=-90)

#saving the figure
fg.figure.savefig("malereg.png")
```

```python
#bar graph for film gender-split vs. film revenue (male actors)
at = sns.barplot(x="percentage male", y="revenue", data=gender1, capsize=.2)

#setting figure size
sns.mpl.rc("figure", figsize=(10, 12))

# plotting graph title and axis labels
plt.title('Comparing the distribution of %of actors who are men in films and film revenue')
plt.xlabel('percentage of actors in films who are men')
plt.ylabel('film revenue:($billions)')
plt.xticks(rotation=-80)

#saving the figure
at.figure.savefig("barchartmale.png")
```
