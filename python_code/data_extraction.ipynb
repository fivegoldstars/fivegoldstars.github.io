```python
import pandas as pd
from IPython.display import display

movies_metadata = pd.read_csv(".\movies_metadata.csv", dtype={'vote_average': float}, low_memory=False)
keywords = pd.read_csv(".\keywords.csv", low_memory=False)

#movies_metadata['id']= movies_metadata['id'].astype(int)


#vote_average is user score from tmdb

#dropping columns we don't want
movies_metadata = movies_metadata.drop(['adult', 'belongs_to_collection', 'homepage', 'original_language','spoken_languages', 'original_title', 'overview', 'poster_path', 'production_countries', 'tagline', 'video' ], axis=1)

#movies_metadata.columns=['movie budget', 'genres', 'id', 'imdb_id', 'popularity', 'production_companies', 'release_data', 'revenue', 'runtime', 'status', 'title', 'vote_average', 'vote_count']

movies_metadata = movies_metadata[movies_metadata['movie budget']!=0]
movies_metadata = movies_metadata[movies_metadata['vote_average']!=0]
movies_metadata = movies_metadata[movies_metadata['revenue']!=0]
movies_metadata = movies_metadata[movies_metadata['status']=='Released']
movies_metadata = movies_metadata[movies_metadata['production_companies']!='[]']

movies_keywords = movies_metadata.merge(keywords, left_index=True, right_index=True)

movies_keywords.tail()
```
